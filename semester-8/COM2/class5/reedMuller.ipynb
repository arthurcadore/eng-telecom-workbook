# Calculate Combination Values
Use a for loop to calculate the combination values for i in range(0, 20) using math.comb(32, i).

```python
# Calculate Combination Values
comb_values = [math.comb(32, i) for i in range(26)]

# Plot the values
plt.plot(range(26), comb_values, marker='o')
plt.xlabel('i')
plt.ylabel('Combination Value')
plt.title('Combination Values for math.comb(32, i)')
plt.grid(True)
plt.show()
```
