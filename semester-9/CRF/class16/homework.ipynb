```python
C1 = 2.5
C2 = 1.5
V1 = 1
V2 = 2

cd0 = 3.5

deltaC = C2 - C1
deltaV = V2 - V1

kc = deltaC / deltaV

c0 = cd0  # c0 is just cd0 minus kc times Vcd, but Vcd is not defined yet, so keep c0 = cd0

# monte um plot da reta variando de 0 a 10 no eixo x (Vcd)
import numpy as np
import matplotlib.pyplot as plt
Vcd = np.linspace(0, 10, 100)  # Mudança aqui: de 0 a 10
Ccd = c0 + kc * Vcd
plt.figure(figsize=(16, 9))
plt.plot(Vcd, Ccd, label='Reta de Ccd', linewidth=2)
plt.xlabel('Vcd (V)')
plt.ylabel('Ccd (pF)')
plt.title('Reta de Ccd - Plano Cartesiano')
plt.xlim(0, 5) # Definindo explicitamente os limites do eixo x
plt.ylim(0, cd0 + 1) # Definindo explicitamente os limites do eixo y
plt.axhline(y=C1, color='orange', linestyle='--', label='C1 = {}'.format(C1))
plt.axhline(y=C2, color='purple', linestyle='--', label='C2 = {}'.format(C2))
plt.axvline(x=V1, color='g', linestyle='--', label='V1 = {}'.format(V1))
plt.axvline(x=V2, color='b', linestyle='--', label='V2 = {}'.format(V2))
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
