```python
e0 = 5 # amplitude da portadora
eM = 2 # amplitude da modulante 

# Para aplicarmos a modulação, também é necessário um indice de modulação. 
# o indice de modulação "m", deve ser menor ou igual á 1 (caso limite para que não ocorra distorção)

# o indice é calculado com base em: 

m = eM / e0
print(f'Indice de modulação: {m}')

# onde: 
# - E0 é a amplitude da portadora 
# - Em é a amplitude da modulante


A = 2*e0 - 2*eM
B = 2*e0 + 2*eM

# desta forma, é possivel também calcular utilizando: 

m = (B-A) / (B+A)
print(f'Indice de modulação 2: {m}')
```

```python
#portência da portadora: 

p0 = e0**2/2

# potência bandas laterais (BLS e BLI): 

pB = ((m**2) * (e0**2))/8

pM_AM_DSB = (e0**2/2) * 2*pB

# sendo: 
# - pB: potência distribuida nas bandas laterais. 
# - (e0**2/2): potência da portadora. 

print(f'Potência das bandas laterais: {pB}')
print(f'Potência da portadora: {pM_AM_DSB}')

```

```python
# Eficiência da modulação AM:

nP = m**2 / (2+ m**2)

fator_nP = nP * 100

print(f'Eficiência da modulação AM: {fator_nP} %')
```
