```python
from math import pow
from scipy.optimize import newton

print("== Questão 1 ==")
PV1 = 100000
i1 = 0.0085
n1 = 60
P1 = PV1 * (i1 * pow(1 + i1, n1)) / (pow(1 + i1, n1) - 1)
print(f"Prestação mensal (PRICE): R$ {P1:.2f}")

print("\n== Questão 2 ==")
P2 = 1000
i2 = 0.01
n2 = 6
PV2 = P2 * (1 - pow(1 + i2, -n2)) / i2
print(f"Valor presente (empréstimo): R$ {PV2:.2f}")

print("\n== Questão 3 ==")
VF3 = 1000000
i3 = 0.0275 / 12
n3 = 35 * 12
P3 = (VF3 * i3) / (pow(1 + i3, n3) - 1)
print(f"Poupança mensal necessária: R$ {P3:.2f}")

print("\n== Questão 4 ==")
PV4 = 50000 * 0.75
PMT4 = 999
n4 = 48
def f4(i):
    return PMT4 * (1 - pow(1 + i, -n4)) / i - PV4
i4 = newton(f4, 0.01)
print(f"Taxa de juros mensal aproximada: {i4 * 100:.4f}%")

print("\n== Questão 5 ==")
VF5 = 30000
i5 = 0.07 / 12
n5 = 24
P5 = (VF5 * i5) / (pow(1 + i5, n5) - 1)
print(f"Depósito mensal necessário: R$ {P5:.2f}")

print("\n== Questão 6 ==")
# Opção 1
entrada_1 = 50000 * 0.25
parcela_1 = 987.52
total_parcelas_1 = parcela_1 * 48
total_pago_1 = entrada_1 + total_parcelas_1

# Opção 2
entrada_2 = 50000 * 0.50
parcela_2 = 1173.34
total_parcelas_2 = parcela_2 * 24
total_pago_2 = entrada_2 + total_parcelas_2

print(f"Total pago na Opção 1: R$ {total_pago_1:.2f}")
print(f"Total pago na Opção 2: R$ {total_pago_2:.2f}")
print("Melhor opção:", end=" ")
if total_pago_1 < total_pago_2 and total_pago_1 < 50000:
    print("Opção 1")
elif total_pago_2 < total_pago_1 and total_pago_2 < 50000:
    print("Opção 2")
else:
    print("À vista")

print("\n== Questão 7 ==")
PV7 = 200000
i7 = 0.01
n7 = 120

# SAC
A7 = PV7 / n7
J1_SAC = PV7 * i7
P1_SAC = A7 + J1_SAC

# PRICE
P7 = PV7 * (i7 * pow(1 + i7, n7)) / (pow(1 + i7, n7) - 1)

print(f"SAC -> Amortização: R$ {A7:.2f}, Juros 1ª: R$ {J1_SAC:.2f}, Prestação 1ª: R$ {P1_SAC:.2f}")
print(f"PRICE -> Prestação mensal: R$ {P7:.2f}")
```
